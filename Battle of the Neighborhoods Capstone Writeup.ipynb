## 1. Introduction

### 1.1 Background

In a city like Toronto, ON, one of the biggest metropolitan centers in Canada, population density is high and entertainment and dining experiences are in demand. However, there are already a large number of options for customers to dine at. While there is a high demand for restaurants, there are also a number of competitors that already exist. Opening a restaurant is a costly endeavour. It is also important to understand that even restaurants offering the best food and service can fail for other reasons. For potential restaurant owners, the success of their business venture will rely on consideration of population and income of the target demographics in the neighbourhoods as well as consideration of the existing competitors that the company may face. 

 ### 1.2 Problem Statement

***If someone is looking to open a restaurant in Toronto, what neighborhoods would be best for them to open it?***

If a restaurant owner were to open a restaurant that is in a low-population area, the brand may not become widly recognized or attended by locals or attract people from other areas. If a restaurant is open in an area with low-income the restaurant will again not likely have many customers or frequency of repeat visits even if the population is there as the customer base may not be able to afford dinning out. Combining low-income and low-population would be an aggrivated scenario affecting the ability for the restaurant to continue operations. Moreover, even in well populated and good income neighbourhoods, if there are a lot of well-known restaurant options already existing, customer loyalty might be strong and the new openning ignored. All of these factors are important to consider in a new venture as they directly impact profitability and the possible longevity of the restaurants existence. Therefore, for potential restaurant owners, the problem of location is worth paying attention to and doing some analysis before even starting the search for a venue. 

### 1.3 Audience

The target audience for this analysis is prospective restaurant owners. Though experienced owners can benefit from this project, new prospective restaurant owners without name recognition may find this information the most beneficial. Another potential stakeholder for this project are small business owners. Since small businesses success can also rely on the population and income of the retaillocation, similar to restaurants, demographics of the neighbourhood they select can be imperative to success. ## Methodology

The first step was to clean the data. The population and income data needed to be assigned into a table of Toronto neighbourhoods and changed from object to numbers for calculation purposes. Once the data was compiled into a table, it came to manipulating the data and using visualization to determine what kind of trends in neighbourhood location to income/population could be derived. This step took many different sections, focusing on the income and population of neighbourhoods each individually, and together represented by the total value aproximation. The total value of a neighbourhood was achieved by multiplying the population by the income. 

Once the population and income data was manipulated to come up with the top neighbourhoods with the highest possible profitability, the Foursquare data was called in to review how these neighbourhood profiles compared based on existing restaurants. This data was viewed on a map to understand where the greatest concentration of restaurants existed. This was further evaluated to determine how many restaurants per people exist in each neighbourhood.

## Discussion

The first point to note is that although we saw that certain neighbourhoods had higher income, given there are less people living in those neighbourhoods and they are somewhat removed from passer-buyers of lower income brackets, they may not be the best location. This is because, although those residents have high discretionarry capital, they would not likely eat out every night and this makes the relative spending per restaurant in the area lower. Next, when looking at the highest number of residents we can see that population alone is not the strongest metric either. Many of the higher density areas have a lower income bracket which would discourage spending on dinning experiences frequently. From this, we can determine that there is a "sweet spot" of strong relative per-household income, as well as population density. The top rated neighbourhoods are those which have the strongest overall value as determined by the income multiplied by the population. In this way, a restaurant would achieve the optimal number of potential customers who are willing to spend money. 

Based on the cluster analysis of the data presented above, we can see that althoough certain neighbourhoods met our top total value criteria of having strong levels of population and income, certain areas have much higher numbers of restaurants already. These are neighbourhoods to avoid since the competition is feirce and would make attracting new customers challenging. As a result, if we compare the neighbourhoods with mid-low restaurant presence already existing we can determine the neighbourhoods that are not already over-serviced by restaurants and have room for noticable restaurants to set up and stand out.
